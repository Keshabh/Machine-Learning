```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading sonar data
df = pd.read_csv('/content/SonarData.csv', header=None)
df.head()
```

```python
df[60].value_counts()
```

```python
#lets group it on the basis of dependent feature,and see the mean values of other columns
df.groupby(60).mean()
```

```python
#lets seprate dependent and independent features
Y = df[[60]]
Y.head()
```

```python
X = df.drop(columns = [60], axis=1)
X.head()
```

```python
#now lets split the data into train, test
#startify  = y makes sure both class A, B is distributed in same proprtion in both train-test
#i.e A: B in train is 90:10, then also in test, it should be 90:10
#random_state = xyz, where xyz can be any value, it just provides randomness in the data, and makes sure everytime this code is run, it returns same random data
x_train, x_test, y_train, y_test = train_test_split(X,Y,random_state=2, test_size=0.1,stratify = Y)
```

```python
print(X.shape, x_train.shape, x_test.shape)
```

```python
#model training
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
#lets predict on training data first
x_train_prediction = model.predict(x_train)
accuracy_score(x_train_prediction, y_train)
```

```python
#lets see it now on test data
x_test_prediction = model.predict(x_test)
accuracy_score(x_test_prediction, y_test)
```
