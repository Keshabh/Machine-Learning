```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/diabetes.csv')
df.head()
```

```python
#here, mean values range of each feature is in very contrast, we need to standardize them to bring them  in the same range
#otehrwise, the impact of high valued feature will override the lower valued feature, and can bias the result.
#bias the result, means, result will then rely more on the basis of high valued feature.
#and will disregard the low valued feature.
df.count()
```

```python
#lets see for each 0 and 1, mean value for each feature
df.groupby('Outcome').mean()
```

```python
#we dont have any null values, so we dont need to do any null handling here
```

```python
#lets split the data into dependent and independent features
x = df.drop(columns='Outcome',axis=1)
x.head()
```

```python
y = df[['Outcome']]
y.head()
```

```python
#lets perform the standardization on x data
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

#this x_sclaed is in numpy dataformat, we need to convert this back into dataframe
x_transformed = pd.DataFrame(x_scaled, columns = x.columns)
x_transformed.head()
```

```python
x_transformed.describe()
```

```python
#now the values are so small, and they are almost in the same range.
#now lets perform train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state=3, stratify=y)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn import svm
from sklearn.metrics import accuracy_score
classifier  = svm.SVC(kernel='linear')
classifier.fit(x_train, y_train)
```

```python
#lets evaluate the model now
y_pred = classifier.predict(x_test)
accuracy_score(y_pred, y_test)
```

```python
#lets see the accuracy on train data as well
y_train_pred = classifier.predict(x_train)
accuracy_score(y_train_pred, y_train)
```

```python
#here, we can see that model is performing well on the training data, but not much well on test data
#it seems to be the case of overfitting scenario.

#lets make a predictive system
#lets pass a data of diabetes and check what does it generate
input_data = (9,152,78,34,171,34.2,0.893,33)
data = np.array(input_data)
data = data.reshape(1,-1)
print(data)
#model is trained on standardized data, so this input data also needs to be standardized
data = scaler.transform(data)
print(data)
data = pd.DataFrame(data, columns = x.columns)
data
```

```python
classifier.predict(data)
```

```python
#since actual output is also 1, and it has also predicted 1 only
#lets see the same for non-diabetic case as well
data = (0,135,94,46,145,40.6,0.284,26)
data = np.array(data)
data = data.reshape(1,-1)
data = scaler.transform(data)
classifier.predict(data)
```
