### Problems

```python
#Problem 1: Email domain counter
text= '''alice@gmail.com
bob@yahoo.com
carol@gmail.com
dave@outlook.com
eve@yahoo.com'''

emailMap = {}
for email in text.split('\n'):
  k = email.split('@')[1]
  if k in emailMap:
    emailMap[k] += 1
  else:
    emailMap[k] = 1

for k,v in emailMap.items():
  print(k,v)
```

```python
#Problem 2: Top 3 words
text = 'This is a test. This test is only a test.'
text = text.replace('.','')
wordMap =  {}
for word in text.split(' '):
  if word in wordMap:
    wordMap[word]+=1
  else:
    wordMap[word]=1

for word, count in sorted(wordMap.items(), key = lambda x:x[1], reverse=True)[:3]:
  print(word, count)
```

# Module 3

### DataFrame

```python
# Read from csv file
# df = pd.read_csv('data.csv')

#create a DataFrame from a dictionary
students = {
    "Alice": {"age": 30, "score": 90},
    "Bob": {"age": 25, "score": 85},
    "Carol": {"age": 25, "score": 92},
    "Dave": {"age": 30, "score": 85}
}

import pandas as pd
df = pd.DataFrame(students)
print(df.head())

#Data Selection
print(df['Alice'])

#Multiple column selection
print(df[['Alice', 'Bob']])

#Loc
print(df.loc['age','Alice'])

#iloc
print(df.iloc[0,2:4])

#Masking
df[df['Alice']<25]
```

### Cleaning Data

```python
# Drop null values
df.dropna()

#FIll na with some value
df.fillna(10)

#Rename columns
df.rename(columns = {'Alice': 'Col1', 'Bob': 'Col2', 'Carol': 'Col3', 'Dave':'Col4'}, inplace=True)

#Replace values
df.replace({'Col1':{30:40}}, inplace=True)

#Change Data Type
df['Col1'] = df['Col1'].astype(float)
```

### Aggregation

```python
#Group BY
df.groupby('Col1')['Col2'].sum()

#Aggregate multiple columns
df.groupby('Col1').agg({'Col2': 'sum', 'Col3': 'mean'})
```

### Sorting

```python
#sort by Col1 ascending, and then Col2 descneding
df.sort_values(by=['Col1', 'Col2'], ascending=[True, False])
```

### Merge

```python
df1 = pd.DataFrame({
    'student_id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Carol', 'Dave'],
    'age': [20, 21, 22, 23]
})

df2 = pd.DataFrame({
    'student_id': [1, 2, 4, 5],
    'score': [88, 92, 79, 85]
})

pd.merge(df1, df2, on="student_id", how="inner")
```

### Problems

```python
#Problem1: Find total sales by region
data = {
    'Region': ['East', 'West', 'East', 'North', 'West'],
    'Sales': [200, 300, 150, 400, 250]
}

df = pd.DataFrame(data)
df.groupby('Region')['Sales'].sum()
```

```python
#Problem 2: Top Earners Per Department
data = [
    {'Name': 'Alice', 'Department': 'IT', 'Salary': 90000},
    {'Name': 'Bob', 'Department': 'HR', 'Salary': 60000},
    {'Name': 'Carol', 'Department': 'IT', 'Salary': 95000},
    {'Name': 'Dave', 'Department': 'Finance', 'Salary': 70000},
    {'Name': 'Eve', 'Department': 'Finance', 'Salary': 85000}
]

df = pd.DataFrame(data)
#here transform acts as a window function, and adds max value in each record of that grouped column i.e Department, and further can be used as a mask
#to select the one that matches with it.
mask = df.groupby('Department')['Salary'].transform('max')
df[df['Salary']== mask]
```

```python
#Problem 3: Filter by Year 2024
data = {
    'Name': ['Alice', 'Bob', 'Carol', 'Dave'],
    'JoinDate': ['2023-05-01', '2024-01-15', '2024-07-20', '2022-11-11']
}

df = pd.DataFrame(data)
df[df['JoinDate'].str.startswith('2024')]
```

```python
###DateTime
#if there is a column in date format, always first convert it into dateTime format as
import pandas as pd
df = pd.DataFrame({
    "date": ["2025-09-04 15:30:45", "2025-01-01 08:15:00"]
})

df['newCol'] = pd.to_datetime(df['date'])
print(df['newCol'].dt.strftime('%Y-%m-%d'))
print(df['newCol'].dt.day)
```
