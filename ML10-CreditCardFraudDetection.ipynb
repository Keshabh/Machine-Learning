```python
#here 1 is for fraud transaction, and 0 represent legit transaction
#we need to see some statistical difference for each columns between these 2 types of data
legit_transaction = df[df.Class == 0]
fraud_transaction = df[df.Class == 1]

legit_transaction.describe()
```

```python
#lerts group by on Class and see the mean of each columns and see how conrtrast the differences are
df.groupby('Class').mean()
```

```python
#lets do undersampling, which ,means we have reduce the number of records from the majority in the case the data is biased.
legit_transaction_sampled = legit_transaction.sample(492)
legit_transaction_sampled.shape
```

```python
#noe lets combine both
df_data = pd.concat([legit_transaction_sampled, fraud_transaction], axis=0)
df_data.head()
```

```python
df_data.shape
```

```python
df_data.Class.value_counts()
```

```python
X = df_data.drop(columns='Class', axis=1)
Y = df_data['Class']
X.head()
```

```python
Y.head()
```

```python
#splitting the data into train and test
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
#lets train the model
model= LogisticRegression()
model.fit(x_train, y_train)
model
```

```python
#lets see the accruacy score on training data
from sklearn.metrics import accuracy_score
x_train_prediction = model.predict(x_train)
print("Accuracy on Training Data: ", accuracy_score(x_train_prediction, y_train))

#lets do the same on test data
x_test_prediction = model.predict(x_test)
print("Accuracy on Test Data: ", accuracy_score(x_test_prediction, y_test))
```
